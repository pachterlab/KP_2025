```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Mias et. al. DE genes

```python
hourly_de = pd.read_csv('mias_results/mias_hourly_genes.csv')
daily_de = pd.read_csv('mias_results/mias_daily_genes.csv')
daily_de.head()
```

```python
import re

# Function to extract item name from the string
def extract_item_name(cell):
    if isinstance(cell, str):
        match = re.search(r'\{(["\']?)(.*?)\1,', cell)
        if match:
            return match.group(2)
    return None

# Apply the function to all cells in the DataFrame
cleaned_df = hourly_de.map(extract_item_name)
cleaned_daily_df = daily_de.map(extract_item_name)
# Display the cleaned DataFrame
cleaned_df.head()
```

## Import sleuth DE genes

```python
hourly_spline = pd.read_csv('hourly_de_new_filter.csv', index_col=0)
daily_spline = pd.read_csv('daily_de_new_filter.csv', index_col=0)
```

## Find overlap in DE genes
Finding overlap in DE genes between sleuth results and Mias et. al. results

```python
def find_name(df1, df2):
    results = []
    
    for name in df2['ext_gene']:
        found_in = []
        for column in df1.columns:
            if name in df1[column].values:
                found_in.append(column)
        if found_in:
            results.append((name, found_in))
        else:
            #results.append((name, []))
            continue
    
    return pd.DataFrame(results, columns=['Gene', 'Found_in_Columns'])

# Run the function and get results
result_df = find_name(cleaned_df, hourly_spline)
print(result_df)
```

```python
daily_result_df = find_name(cleaned_daily_df, daily_spline)
print(daily_result_df)
```

```python
hourly_cluster_counts = []
for cluster in result_df['Found_in_Columns']:
    for item in cluster:
        hourly_cluster_counts.append(item)

hourly_count_series = pd.Series(hourly_cluster_counts).value_counts()
hourly_categories = cleaned_df.columns

# Reindex the series to include all categories, filling missing ones with 0
hourly_count_series = hourly_count_series.reindex(hourly_categories, fill_value=0)
num_overlap = len(result_df['Gene'].unique())

# Plot the bar chart
hourly_count_series.plot(kind='bar', color='skyblue', figsize=(8, 6))

# Add labels and title
plt.xlabel('Cluster from Original Publication')
plt.ylabel('Count')
plt.title('Hourly DE genes commonly identified in current and original study')
plt.text(7, 78, f'Total # commonly DE genes: {num_overlap}')
plt.savefig('hourly_overlap.png', dpi=300, bbox_inches='tight')
```

```python
daily_cluster_counts = []
for cluster in daily_result_df['Found_in_Columns']:
    for item in cluster:
        daily_cluster_counts.append(item)

count_series = pd.Series(daily_cluster_counts).value_counts()
categories = cleaned_daily_df.columns

count_series = count_series.reindex(categories, fill_value=0)
num_overlap = len(daily_result_df['Gene'].unique())

# Plot the bar chart
count_series.plot(kind='bar', color='skyblue', figsize=(8, 6))

# Add labels and title
plt.xlabel('Cluster from Original Publication')
plt.ylabel('Count')
plt.title('Daily DE genes commonly identified in current and original study')
plt.text(10, 80, f'Total # commonly DE genes: {num_overlap}')
plt.savefig('daily_overlap.png', dpi=300, bbox_inches='tight')
```
